# Advanced Lane Finding

## Goals / steps of the project

- ~~Compute the camera calibration matrix and distortion coefficients given a set of chessboard images.~~
- ~~Apply a distortion correction to raw images.~~

### Compute the camera calibration matrix and distortion coefficients given a set of chessboard images

```python
## Object Points class
## Stores object points and image points
class ObjectPoints:
    def __init__(self, nx, ny):
        self.nx = nx
        self.ny = ny
        # Prepare object points
        self.objp = np.zeros((nx*ny,3), dtype=np.float32)
        self.objp[:,:2] = np.mgrid[0:nx, 0:ny].T.reshape(-1,2)
        # Arrays to store objectpoints and imagepoints
        self.objectpoints = []
        self.imagepoints = []
        # Undistort values
        self.ret = False
        self.mtx = False
        self.dist = False
        self.rvecs = False
        self.tvecs = False
    
    def getObjectPoints(self):
        return self.objectpoints
    
    def getImagePoints(self):
        return self.imagepoints
    
    def getReturnVal(self):
        return self.ret
    
    def getMatrix(self):
        return self.mtx
    
    def getDist(self):
        return self.dist
    
    def getRvecs(self):
        return self.rvecs
    
    def getTvecs(self):
        return self.tvecs
    
    def addObjectPoint(self):
        self.objectpoints.append(self.objp)
        
    def addImagePoint(self, imagepoints):
        self.imagepoints.append(imagepoints)
        
    def calibrate(self, img_size):
        self.ret, self.mtx, self.dist, self.rvecs, self.tvecs = cv2.calibrateCamera(self.objectpoints, self.imagepoints, img_size, None, None)
```

```python
## Detect corners function
def detectCorners(img, objp, nx=9, ny=6):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    retval, corners = cv2.findChessboardCorners(gray, (nx, ny), None)
    if retval == True:
        objp.addObjectPoint()
        objp.addImagePoint(corners)
        cv2.drawChessboardCorners(img, (nx, ny), corners, retval)
    return img
```

### Use color transforms, gradients, etc., to create a thresholded binary image.

```python
from ipywidgets import interact, interactive, fixed, interact_manual

## Absolute Sobel threshold
def abs_sobel_threshold(img, orient='x', sobel_kernel=3, thresh=(0, 255)):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    x = 1 if orient == 'x' else 0 # Set x to 1 if orientation is x
    y = 1 if orient == 'y' else 0 # Set y to 1 if orientation is y
    sobel = np.absolute(cv2.Sobel(gray, cv2.CV_64F, x, y, ksize=sobel_kernel))
    scale = np.max(sobel) / 255
    sobel = (sobel / scale).astype(np.uint8)
    grad_binary = np.zeros_like(sobel)
    grad_binary[(sobel >= thresh[0]) & (sobel <= thresh[1])] = 1
    return grad_binary

## Magnitude threshold
def magnitude_threshold(img, sobel_kernel=3, mag_thresh=(0, 255)):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
    sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
    grad = np.sqrt(sobel_x ** 2 + sobel_y ** 2)
    scale = np.max(grad) / 255
    grad = (grad / scale).astype(np.uint8)
    mag_binary = np.zeros_like(grad)
    mag_binary[(grad >= mag_thresh[0]) & (grad <= mag_thresh[1])] = 1
    return mag_binary

## Direction threshold
def direction_threshold(img, sobel_kernel=3, thresh=(0, np.pi/2)):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    sobel_x = np.absolute(cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel))
    sobel_y = np.absolute(cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel))
    grad = np.arctan2(sobel_y, sobel_x).astype(np.uint8)
    dir_binary = np.zeros_like(grad)
    dir_binary[(grad >= thresh[0]) & (grad <= thresh[1])] = 1
    return dir_binary

## Color threshold
def color_mask(img, thresh):
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(hsv, thresh[0], thresh[1])
    return mask

def color_threshold(img):
    yellow = color_mask(img, (np.array([0, 100, 100]), np.array([80, 255, 255])))
    white = color_mask(img, (np.array([0, 0, 150]), np.array([255, 80, 255])))
    return cv2.bitwise_or(yellow, white)

def saturation_threshold(img, cthresh):
    hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    s_channel = hls[:,:,2]
    s_binary = np.zeros_like(s_channel)
    s_binary[(s_channel >= cthresh[0]) & (s_channel <= cthresh[1])] = 1
    return s_binary
    
def pipeline(img, ksize=3, mthresh=(30,100), dthresh=(20,100), cthresh=(170,255)):
    magnitude = magnitude_threshold(img, sobel_kernel=ksize, mag_thresh=mthresh)
    direction = abs_sobel_threshold(img, 'x', sobel_kernel=ksize, thresh=dthresh)
    color = color_threshold(img)
    saturation = saturation_threshold(img, cthresh)
    combined = np.zeros_like(direction).astype(np.uint8)
    combined[(saturation == 1) & (color >= .5) | ((direction == 1) & (magnitude == 1))] = 1
    return combined
```

### Detect lane pixels and fit to find the lane boundary.

```python
def find_base(img):
    histogram = np.sum(img[img.shape[0]//2:,:], axis=0)
    midpoint = np.int(histogram.shape[0]//2)
    leftx_base = np.argmax(histogram[:midpoint])
    rightx_base = np.argmax(histogram[midpoint:]) + midpoint
    return (histogram, midpoint, leftx_base, rightx_base)
```

```python
def find_lanes_initial(binary, leftx_base, rightx_base, nwindows=9, margin=100, minpix=50, visualize=True):
    window_height = np.int(binary.shape[0] / nwindows)
    nonzero = binary.nonzero()
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    left_lane_inds = []
    right_lane_inds = []
    leftx_current = leftx_base
    rightx_current = rightx_base
    out_img = np.dstack((binary, binary, binary))*255
    
    for window in range(nwindows):
        win_left = (
            (leftx_current - margin, binary.shape[0] - (window + 1) * window_height),
            (leftx_current + margin, binary.shape[0] - window * window_height)
        )
        win_right = (
            (rightx_current - margin, binary.shape[0] - (window + 1) * window_height),
            (rightx_current + margin, binary.shape[0] - window * window_height)
        )
        if visualize:
            cv2.rectangle(out_img, win_left[0], win_left[1], (0,255,0), 2) 
            cv2.rectangle(out_img, win_right[0], win_right[1], (0,255,0), 2) 
        good_left_inds = ((nonzeroy >= win_left[0][1]) & (nonzeroy < win_left[1][1]) & (nonzerox >= win_left[0][0]) & (nonzerox < win_left[1][0])).nonzero()[0]
        good_right_inds = ((nonzeroy >= win_right[0][1]) & (nonzeroy < win_right[1][1]) & (nonzerox >= win_right[0][0]) & (nonzerox < win_right[1][0])).nonzero()[0]
        left_lane_inds.append(good_left_inds)
        right_lane_inds.append(good_right_inds)
        if len(good_left_inds) > minpix:
            leftx_current = np.int(np.mean(nonzerox[good_left_inds]))
        if len(good_right_inds) > minpix:
            rightx_current = np.int(np.mean(nonzerox[good_right_inds]))
            
    left_lane_inds = np.concatenate(left_lane_inds)
    right_lane_inds = np.concatenate(right_lane_inds)
    
    leftx = nonzerox[left_lane_inds]
    lefty = nonzeroy[left_lane_inds] 
    rightx = nonzerox[right_lane_inds]
    righty = nonzeroy[right_lane_inds] 

    if len(leftx) > 0 and len(rightx) > 0:
        left_fit = np.polyfit(lefty, leftx, 2)
        right_fit = np.polyfit(righty, rightx, 2)

        ploty = np.linspace(0, binary.shape[0]-1, binary.shape[0] )
        left_fitx = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
        right_fitx = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]
        out_img[nonzeroy[left_lane_inds], nonzerox[left_lane_inds]] = [255, 0, 0]
        out_img[nonzeroy[right_lane_inds], nonzerox[right_lane_inds]] = [0, 0, 255]
        fig = Figure(figsize=(16,9), dpi=80)
        canvas = FigureCanvas(fig)
        ax = fig.add_axes((0,0,1,1)) 
        ax.imshow(out_img)
        ax.plot(left_fitx, ploty, color='yellow')
        ax.plot(right_fitx, ploty, color='yellow')
        canvas.draw()
        vision = np.fromstring(canvas.tostring_rgb(), dtype='uint8').reshape(fig.canvas.get_width_height()[::-1] + (3,))

        if visualize:
            plt.imshow(vision)

        return ((leftx, lefty), (rightx, righty), left_fit, right_fit, vision)
    else:
        return ((False, False), (False, False), False, False, False)
```

```python
def track_lanes(binary, left_fit, right_fit, margin=100, visualize=True):
    nonzero = binary.nonzero()
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    margin = 100
    left_lane_inds = ((nonzerox > (left_fit[0]*(nonzeroy**2) + left_fit[1]*nonzeroy + left_fit[2] - margin)) & (nonzerox < (left_fit[0]*(nonzeroy**2) + left_fit[1]*nonzeroy + left_fit[2] + margin))) 
    right_lane_inds = ((nonzerox > (right_fit[0]*(nonzeroy**2) + right_fit[1]*nonzeroy + right_fit[2] - margin)) & (nonzerox < (right_fit[0]*(nonzeroy**2) + right_fit[1]*nonzeroy + right_fit[2] + margin)))  

    leftx = nonzerox[left_lane_inds]
    lefty = nonzeroy[left_lane_inds] 
    rightx = nonzerox[right_lane_inds]
    righty = nonzeroy[right_lane_inds]
    
    # Fit a second order polynomial to each
    left_fit = np.polyfit(lefty, leftx, 2)
    right_fit = np.polyfit(righty, rightx, 2)
    
    ploty = np.linspace(0, binary.shape[0]-1, binary.shape[0] )
    left_fitx = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
    right_fitx = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]
    out_img = np.dstack((binary, binary, binary))*255
    window_img = np.zeros_like(out_img)
    
    # Sanity check, compare slopes
    slope_left = binary.shape[1] / (left_fitx[0] - left_fitx[-1])
    slope_right = binary.shape[1] / (right_fitx[0] - right_fitx[-1])
    if abs(slope_left - slope_right) > .9:
        return((False, False), (False, False), False, False, False)

    # Color in left and right line pixels
    out_img[nonzeroy[left_lane_inds], nonzerox[left_lane_inds]] = [255, 0, 0]
    out_img[nonzeroy[right_lane_inds], nonzerox[right_lane_inds]] = [0, 0, 255]

    # Generate a polygon to illustrate the search window area
    # And recast the x and y points into usable format for cv2.fillPoly()
    left_line_window1 = np.array([np.transpose(np.vstack([left_fitx-margin, ploty]))])
    left_line_window2 = np.array([np.flipud(np.transpose(np.vstack([left_fitx+margin, ploty])))])
    left_line_pts = np.hstack((left_line_window1, left_line_window2))
    right_line_window1 = np.array([np.transpose(np.vstack([right_fitx-margin, ploty]))])
    right_line_window2 = np.array([np.flipud(np.transpose(np.vstack([right_fitx+margin, ploty])))])
    right_line_pts = np.hstack((right_line_window1, right_line_window2))

    # Draw the lane onto the warped blank image
    cv2.fillPoly(window_img, np.int_([left_line_pts]), (0,255, 0))
    cv2.fillPoly(window_img, np.int_([right_line_pts]), (0,255, 0))
    result = cv2.addWeighted(out_img, 1, window_img, 0.3, 0)
    fig = Figure(figsize=(16,9), dpi=80)
    canvas = FigureCanvas(fig)
    ax = fig.add_axes((0,0,1,1)) 
    ax.imshow(result)
    ax.plot(left_fitx, ploty, color='yellow')
    ax.plot(right_fitx, ploty, color='yellow')
    canvas.draw()
    vision = np.fromstring(canvas.tostring_rgb(), dtype='uint8').reshape(fig.canvas.get_width_height()[::-1] + (3,))
    
    if visualize:
        plt.imshow(vision)
        
    return ((leftx, lefty), (rightx, righty), left_fit, right_fit, vision)
            
```

```python
test_image_lanes_org = cv2.imread('./test_images/test3.jpg')
test_image_lanes = undistortImage(test_image_lanes_org)
test_image_lanes = pipeline(test_image_lanes, ksize=9, mthresh=(30,100), dthresh=(100,200), cthresh=(80,255))
test_image_lanes = cv2.warpPerspective(test_image_lanes, M, (test_image_lanes.shape[1], test_image_lanes.shape[0]), flags=cv2.INTER_LINEAR)
test_image_lane_base = test_image_lanes

# Find base of the lane
histogram, midpoint, leftx_base, rightx_base = find_base(test_image_lane_base)

plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.title("Image processed")
plt.imshow(test_image_lanes, cmap='gray')
plt.subplot(1,2,2)
plt.title("Histogram")
plt.plot(histogram)
plt.figure(figsize=(16,10))
plt.title("Initially found lanes")

# Initially find lanes, using the base
left, right, left_fit, right_fit, img_initial = find_lanes_initial(test_image_lane_base, leftx_base, rightx_base)

plt.figure(figsize=(16,9))
plt.title("Track lanes")

# Lanes found so keep track of them
left, right, left_fit, right_fit, img_track = track_lanes(test_image_lane_base, left_fit, right_fit)
```

### Determine the curvature of the lane and vehicle position with respect to center.

```python
def calculateCurve(coords, y):
    ym_per_pix = 30/720
    xm_per_pix = 3.7/700
    new_fit = np.polyfit(coords[1] * ym_per_pix, coords[0] * xm_per_pix, 2)
    return ((1 + (2 * new_fit[0] * y + new_fit[1]) ** 2) ** 1.5) / np.absolute(2 * new_fit[0])

def curvature(left, right, y=None):
    if y is None:
        y = 720
    return (calculateCurve(left, y), calculateCurve(right, y))
```

```python
cLeft, cRight = curvature(left, right)
print("CURVATURE LEFT LANE = {:.3f} M".format(cLeft))
print("CURVATURE RIGHT LANE = {:.3f} M".format(cRight))
```

```python
def offset_middle(left, right):
    xm_per_pix = 3.7/700
    left_val = left[0][-1]
    right_val = right[0][-1]
    width = right_val - left_val
    return (1280 / 2 - (left_val + width / 2)) * xm_per_pix
```

```python
print("OFFSET MIDDLE = {:.3f} M".format(offset_middle(left, right)))
```

### Warp the detected lane boundaries back onto the original image.

```python
def draw_line(img, line, color=(255,0,0)):
    p = np.poly1d(line)
    x = list(range(0, img.shape[0]))
    y = list(map(int, p(x)))
    pts = np.array([[_y,_x] for _x, _y in zip(x, y)])
    pts = pts.reshape((-1,1,2))
    cv2.polylines(img, np.int32([pts]), False, color=color, thickness=10)
    return pts
    
def visualize(img, left, right, color=(0,150,0)):
    ## Visualize lanes
    out_image = np.zeros_like(img)
    left_pts = draw_line(out_image, left, (255,0,0))
    right_pts = draw_line(out_image, right, (0, 0, 255))
    top = [left_pts[-1], right_pts[-1]]
    bottom = [left_pts[0], right_pts[0]]
    cv2.fillPoly(out_image, [np.concatenate((right_pts, left_pts, top, bottom))], color=color)
    
    ## Warp lanes back to normal space
    out_image = cv2.warpPerspective(out_image, Minv, (out_image.shape[1], out_image.shape[0]), flags=cv2.INTER_LINEAR)
    
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    ## Combine images
    final = cv2.addWeighted(img_rgb, 0.7, out_image, 1, .1)
    return final
```

### Output visual display of the lane boundaries and numerical estimation of lane curvature and vehicle position.

```python
## Definite processing of image
class Line:
    def __init__(self):
        self.detected = False  
        self.recent_xfitted = []
        self.last_fit = False
        self.current_fit = False  
        self.coeff = []
        self.coords = []
        self.lastImage = False
    
    def getDetected(self):
        return self.detected
    
    def setDetected(self, value):
        self.detected = value
    
    def getCurrentFit(self):
        if self.last_fit is not False:
            return self.last_fit * 0.05 + self.current_fit * 0.95
        else:
            return self.current_fit
    
    def setCurrentFit(self, poly):
        self.last_fit = self.current_fit
        self.current_fit = poly
        if (abs(self.last_fit - self.current_fit) > 0.005):
            self.setDetected(False)
    
    def setCoord(self, coord):
        self.coords.append(coord)
        
    def getCoord(self):
        return self.coords[-1]
    
    def setLastImage(self, img):
        self.lastImage = img
        
    def getLastImage(self):
        return self.lastImage
    
left_line = Line()
right_line = Line()

def process_image(image):
    img_org = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    img_org = undistortImage(img_org)
    img_pipeline = pipeline(img_org, ksize=9, mthresh=(30,100), dthresh=(100,200), cthresh=(80,255))
    img = cv2.warpPerspective(img_pipeline, M, (img_org.shape[1], img_org.shape[0]), flags=cv2.INTER_LINEAR)
    org_warped = cv2.warpPerspective(img_org, M, (img_org.shape[1], img_org.shape[0]), flags=cv2.INTER_LINEAR)
    histogram, midpoint, leftx_base, rightx_base = find_base(img)
    
    if left_line.getDetected() is False:
        left, right, left_fit, right_fit, img_lanes = find_lanes_initial(img, leftx_base, rightx_base, visualize=False)
        if left_fit is not False:
            left_line.setDetected(True)
            right_line.setDetected(True)
        else:
            return cv2.cvtColor(org_warped, cv2.COLOR_BGR2RGB)
    else:
        left, right, left_fit, right_fit, img_lanes = track_lanes(img, left_line.getCurrentFit(), right_line.getCurrentFit(), visualize=False)
    
    if left_fit is not False:
        left_line.setCurrentFit(left_fit)
        right_line.setCurrentFit(right_fit)
        left_line.setCoord(left)
        right_line.setCoord(right)
        left_line.setLastImage(img_lanes)
    else:
        left_line.setDetected(False)
        right_line.setDetected(False)
    
    curveLeft, curveRight = curvature(left_line.getCoord(), right_line.getCoord())
    offset = offset_middle(left_line.getCoord(), right_line.getCoord())

    color = (0,150,0) if left_line.getDetected() is True else (0,140,0)
    final = visualize(cv2.cvtColor(image,cv2.COLOR_BGR2RGB), left_line.getCurrentFit(), right_line.getCurrentFit(), color)

    font = cv2.FONT_HERSHEY_COMPLEX

    middlepanel = np.zeros((120, 1280, 3), dtype=np.uint8)
    cv2.putText(middlepanel, 'Estimated lane curvature: {:.3f}'.format(curveLeft), (30, 60), font, 1, (255,0,0), 2)
    cv2.putText(middlepanel, 'Estimated Meters right of center: {:.3f}'.format(offset), (30, 90), font, 1, (255,0,0), 2)

    plt.figure(figsize=(16,9))
    # assemble the screen example
    diagScreen = np.zeros((1080, 1920, 3), dtype=np.uint8)
    diagScreen[0:720, 0:1280] = final
    diagScreen[0:240, 1280:1600] = cv2.resize(cv2.cvtColor(img_org, cv2.COLOR_BGR2RGB), (320,240), interpolation=cv2.INTER_AREA) 
    diagScreen[0:240, 1600:1920] = cv2.resize(cv2.cvtColor(org_warped, cv2.COLOR_BGR2RGB), (320,240), interpolation=cv2.INTER_AREA)
    #diagScreen[240:480, 1280:1600] = cv2.resize(img_lanes, (320,240), interpolation=cv2.INTER_AREA)
    #diagScreen[240:480, 1600:1920] = cv2.resize(img_show, (320,240), interpolation=cv2.INTER_AREA)*4
    diagScreen[600:1080, 1280:1920] = cv2.resize(left_line.getLastImage(), (640,480), interpolation=cv2.INTER_AREA)*4
    diagScreen[720:840, 0:1280] = middlepanel
    diagScreen[840:1080, 0:320] = cv2.resize(final, (320,240), interpolation=cv2.INTER_AREA)
    diagScreen[840:1080, 320:640] = cv2.resize(final, (320,240), interpolation=cv2.INTER_AREA)
    diagScreen[840:1080, 640:960] = cv2.resize(final, (320,240), interpolation=cv2.INTER_AREA)
    diagScreen[840:1080, 960:1280] = cv2.resize(final, (320,240), interpolation=cv2.INTER_AREA)
    return diagScreen
    
```
